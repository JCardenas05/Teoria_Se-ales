```python
def coloredhist(X, plot=None):

    """
    Crea un histograma coloreado en función de la frecuencia de los valores.

    Parámetros:
    - X: Datos a graficar.
    - plot: Índice del gráfico.
    """
    Nbins = int(1 + 3.322*np.log2(len(X)))  # Cálculo del número de bins usando la regla de Sturges
    # Creación del histograma

    if plot != None:
        n, bins, patches = ax[plot].hist(X, bins=Nbins, density=True)
    else:
        n, bins, patches = ax.hist(X, bins=Nbins, density=True)
    # Normalización de las frecuencias para determinar colores
    fracs = n / n.max()
    norm = colors.Normalize(fracs.min(), fracs.max())
    # Asignación de colores según la frecuencia relativa
    for thisfrac, thispatch in zip(fracs, patches):
        color = plt.cm.viridis(norm(thisfrac))
        thispatch.set_facecolor(color)
    plt.grid()  # Añadir rejilla al gráfico

    return
```

```python
# Funciones para calcular la pdf de una distribución normal con media positiva y negativa
def normal_distribution_function_r(x,mean,std):
    value = scipy.stats.norm.pdf(x, mean, std)  # Calcular la pdf con media positiva
    return value

def normal_distribution_function_l(x,mean,std):
    value = scipy.stats.norm.pdf(x, -mean, std)  # Calcular la pdf con media negativa
    return value
```

```python
def norMaxMin(data):
    """
    Normaliza los datos basándose en el valor mínimo y máximo.

    Parámetros:
    - data: Datos a normalizar.

    Retorna:
    - Datos normalizados.
    """
    min_value = np.min(data)  # Encuentra el valor mínimo en los datos
    max_value = np.max(data)  # Encuentra el valor máximo en los datos
    # Normaliza los datos usando los valores mínimo y máximo
    return (data - min_value) / (max_value - min_value)
```

```python
#luego de haber aplicado el método de Cross Validation y verificar su funcionamiento, se identificó que este
# presenta un costo computacional muy alto, por ello se de propone la implementación
# de la regla de Silverman la cual genera resultados mucho mas rápidos pero un poco menos precisos

def kde_function(signal_x,eval_points=None,method="Silverman",res=500):
    """
    Aplica el método de KDE gaussiano a la señal proporcionada y ajusta
    automáticamente el ancho de banda mediante una búsqueda en la cuadrícula.

    Parámetros:
    - signal: Señal a la que se aplica KDE.
    - eval_points: puntos de evaluación
    - method: método

    Retorna:
    - eval_points: Vector de evaluación.
    - y_sp: Función de densidad estimada en los puntos de evaluación.
    """
    if eval_points is None:
        eval_points = np.linspace(np.min(signal_x), np.max(signal_x),res)

    if method == 'Silverman':
        n = len(signal_x)
        sigma = np.std(signal_x)
        optimal_bandwidth = (4 / (3 * n)) ** (1 / 5) * sigma

        print(f'Method:{method}\nOptimal bandwidth: {optimal_bandwidth:.2f}')
        kde = scipy.stats.gaussian_kde(signal_x, bw_method=optimal_bandwidth)
        y_sp = kde.pdf(eval_points)

    elif method == 'Cross Validation':
        # Define los posibles valores para el ancho de banda
        bandwidths = np.arange(0.05, 2, 0.05)

        # Crea y configura el objeto para la búsqueda en la cuadrícula
        kde = KernelDensity(kernel='gaussian')

        grid = GridSearchCV(kde, {'bandwidth': bandwidths})
        grid.fit(signal_x.reshape(-1, 1))
        # Obtiene el mejor estimador y su ancho de banda óptimo
        kde_optimal = grid.best_estimator_
        optimal_bandwidth = kde_optimal.bandwidth
        print(f'Optimal bandwidth: {optimal_bandwidth:.2f}')

        log_dens = kde_optimal.score_samples(eval_points.reshape(-1, 1))
        y_sp = np.exp(log_dens)

    else: return

    return (eval_points, y_sp)
```

Implementación del detector de neyman pearson para el calculo del umbral.


Sí, el cálculo del umbral en el lema de Neyman-Pearson es un aspecto crucial para implementar el detector.

### 1. **Definición de las Hipótesis:**
   - **\(H_0\):** Hipótesis Nula (por ejemplo, la señal recibida es solo ruido).
   - **\(H_1\):** Hipótesis Alternativa (por ejemplo, la señal recibida es una mezcla de señal más ruido).

### 2. **Definición de las Probabilidades a Priori:**
   - Estas son las probabilidades de ocurrencia de cada hipótesis antes de observar la señal.
   - Por ejemplo, si no tenemos ninguna razón para favorecer una hipótesis sobre la otra, podríamos asignar probabilidades iguales a ambas hipótesis.

### 3. **Definición de las Funciones de Verosimilitud:**
   - Estas son las funciones que nos dicen cuán "verosímiles" son los datos observados bajo cada hipótesis.
   - Por lo general, se basan en la distribución probabilística de los datos bajo cada hipótesis.

### 4. **Cálculo de la Razón de Verosimilitudes:**
   - La razón de verosimilitudes se calcula tomando el cociente de la función de verosimilitud de \(H_1\) dividido por la función de verosimilitud de \(H_0\).

### 5. **Definición del Nivel de Significancia (\(\alpha\)):**
   - Es la probabilidad máxima de rechazar la hipótesis nula cuando es verdadera (error de tipo I).

### 6. **Determinación del Umbral (\(\Lambda\)):**
   - Encuentra un umbral \(\Lambda\) tal que la probabilidad de error de tipo I sea igual a \(\alpha\).
   - Esto a menudo implica resolver una ecuación basada en las funciones de verosimilitud y el nivel de significancia elegido.

### 7. **Regla de Decisión:**
   - **Si la razón de verosimilitudes es mayor que \(\Lambda\):** Decidir a favor de \(H_1\).
   - **Si la razón de verosimilitudes es menor o igual que \(\Lambda\):** Decidir a favor de \(H_0\).

### Ejemplo Matemático:
Para ilustrarlo con un ejemplo matemático simple, supongamos que tenemos dos hipótesis con funciones de densidad de probabilidad \(f_0(x)\) y \(f_1(x)\), y queremos controlar el error de tipo I a un nivel \(\alpha\):

\[ P(\text{Rechazar } H_0 | H_0 \text{ es verdadera}) = \alpha \]

Dado un umbral \(\Lambda\), podemos expresar esto como:

\[ \int_{\{x: f_1(x)/f_0(x) > \Lambda\}} f_0(x)dx = \alpha \]

Resolver esta ecuación para \(\Lambda\) nos dará el umbral necesario para el detector de Neyman-Pearson.

### Conclusión:
Calcular el umbral es esencialmente un ejercicio de equilibrio entre las tasas de error, garantizando que el error de tipo I no exceda un cierto nivel mientras intentamos minimizar el error de tipo II. Las formas exactas de las funciones de verosimilitud y la elección del nivel de significancia influirán directamente en la determinación del umbral.


```python
from scipy.stats import gaussian_kde

def neyman_pearson_threshold(S0_pdf, S1_pdf, alpha):
    # Asumimos que S0_pdf[0] y S1_pdf[0] son los valores de x evaluados y S0_pdf[1] y S1_pdf[1] son los valores de la pdf
    # Ordenamos las pdfs para asegurarnos de que estén en el orden correcto
    sorted_indices_S0 = np.argsort(S0_pdf[0])
    sorted_indices_S1 = np.argsort(S1_pdf[0])

    S0_pdf = (S0_pdf[0][sorted_indices_S0], S0_pdf[1][sorted_indices_S0])
    S1_pdf = (S1_pdf[0][sorted_indices_S1], S1_pdf[1][sorted_indices_S1])

    # Inicializamos el umbral en el mínimo valor de x
    threshold = np.min(S0_pdf[0])
    # Calculamos la integral acumulada para la pdf S0
    cumulative_S0 = np.cumsum(S0_pdf[1]) * np.diff(S0_pdf[0], prepend=0)

    # Encontramos el umbral donde el valor acumulado sea igual a (1 - alpha)
    threshold_index = np.where(cumulative_S0 >= (1 - alpha))[0][0]
    threshold = S0_pdf[0][threshold_index]

    return threshold

# Nivel de significancia alpha para Neyman-Pearson (por ejemplo, 0.05)
alpha = 0.05
ideal_threshold = neyman_pearson_threshold(S0_pdf, S1_pdf, alpha)
print(f'Umbral ideal de Neyman-Pearson: {ideal_threshold}')

```
